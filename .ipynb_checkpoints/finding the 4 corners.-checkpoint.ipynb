```python
def scan(input_path , segmented_output):
    
    #semantic segmentation for the id cards
    Model =CreateModel()
    Model.load_weights("best_model.hdf5")
    ImagePath = input_path
    ImageInput= PrepareImage(ImagePath)
    Predicted = np.where(Model.predict(ImageInput)>0.6 , 1,0)

    Predicted=np.array(np.squeeze(Predicted[0]) , dtype='uint8')

    plt.imsave( segmented_output, Predicted, cmap='gray')

    



    image = cv2.imread(segmented_output)
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    thresh = cv2.threshold(gray, 200, 255,cv2.THRESH_BINARY_INV)[1]
    edged = cv2.Canny(thresh, 30, 100)
    # find the largest contour in the threshold image
    cnts = cv2.findContours(edged,cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_SIMPLE)
    
    cnts = imutils.grab_contours(cnts)
    output = image.copy()
    contours = sorted(cnts,key=cv2.contourArea,reverse=True)[:1]
    print(len(contours))
#     area = cv2.contourArea(largest_contour)
#     print(area)
    plt.imshow(edged)
    
    
    for c in contours:
        for eps in np.linspace(0.001, 0.02, 10): 
            p=cv2.arcLength(c,True)
            approx=cv2.approxPolyDP(c,0.02*p,True)
        approx = np.resize(approx , (approx.shape[0] , approx.shape[2]))
        corners = np.zeros((4,2) , dtype = int)
        print(approx.shape)
        target = approx
    #     print(approx)
        if len(approx)==4:
            corners = approx
            break
        elif len(approx) > 4:
            target =np.array(target)
            target_x_sorted = target[target[:, 0].argsort()]
            print(target_x_sorted)


            card_width =  max([q[0] for q in target_x_sorted]) - min([q[0] for q in target_x_sorted])
            card_height = max([q[1] for q in target_x_sorted]) - min([q[1] for q in target_x_sorted])
            j=0
            target_cleaned_x = []
            target_cleaned_y = []
            x_ref = target_x_sorted[0][0]
            for index in target_x_sorted:
                if(index[0] - x_ref < 1/4 * card_width or index[0] - x_ref > 3/4 * card_width):
                    target_cleaned_x.append(index)


            min_x_min_y = min([q[1] for q in target_cleaned_x[:int(len(target_cleaned_x)/2)]])
            min_x_max_y = max([q[1] for q in target_cleaned_x[:int(len(target_cleaned_x)/2)]])
            max_x_min_y = min([q[1] for q in target_cleaned_x[int(len(target_cleaned_x)/2):]])
            max_x_max_y = max([q[1] for q in target_cleaned_x[int(len(target_cleaned_x)/2):]])
            indecies = [np.where(target_cleaned_x == min_x_min_y )[0][0] ,np.where(target_cleaned_x == min_x_max_y )[0][0] 
                       , np.where(target_cleaned_x == max_x_min_y )[0][0] , np.where(target_cleaned_x == max_x_max_y )[0][0]]
            i=0
            for x in indecies:
                corners[i] = target_cleaned_x[x]           
                i+=1
            break
        target_cleaned_x =np.array(target_cleaned_x)


    orig = input_path
    original = cv2.imread(orig)
    warped = four_point_transform(original, corners)
    
        # load dlib's HOG + Linear SVM face detector
    print("[INFO] loading HOG + Linear SVM face detector...")
    detector = dlib.get_frontal_face_detector()
    # load the input image from disk, resize it, and convert it from
    # BGR to RGB channel ordering (which is what dlib expects)
    image = warped
    image = imutils.resize(image, width=600)
    rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    # perform face detection using dlib's face detector
    start = time.time()
    print("[INFO[ performing face detection with dlib...")
    rects = detector(rgb)
    end = time.time()
    print("[INFO] face detection took {:.4f} seconds".format(end - start))

    # convert the resulting dlib rectangle objects to bounding boxes,
    # then ensure the bounding boxes are all within the bounds of the
    # input image
    boxes = [convert_and_trim_bb(image, r) for r in rects]
    # loop over the bounding boxes
    
    if(len(boxes) < 1):
        print('please take the photo again')
        return warped
    else:
        return warped
```
